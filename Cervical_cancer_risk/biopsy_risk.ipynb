```
data_cancer['Number of sexual partners'] = data_cancer['Number of sexual partners'].fillna(data_cancer['Number of sexual partners'].median())
data_cancer['First sexual intercourse'] = data_cancer['First sexual intercourse'].fillna(data_cancer['First sexual intercourse'].median())
data_cancer['Num of pregnancies'] = data_cancer['Num of pregnancies'].fillna(data_cancer['Num of pregnancies'].median())
data_cancer['Smokes'] = data_cancer['Smokes'].fillna(1)
data_cancer['Smokes (years)'] = data_cancer['Smokes (years)'].fillna(data_cancer['Smokes (years)'].median())
data_cancer['Smokes (packs/year)'] = data_cancer['Smokes (packs/year)'].fillna(data_cancer['Smokes (packs/year)'].median())
data_cancer['Hormonal Contraceptives'] = data_cancer['Hormonal Contraceptives'].fillna(1)
data_cancer['Hormonal Contraceptives (years)'] = data_cancer['Hormonal Contraceptives (years)'].fillna(data_cancer['Hormonal Contraceptives (years)'].median())
data_cancer['IUD'] = data_cancer['IUD'].fillna(0) 
data_cancer['IUD (years)'] = data_cancer['IUD (years)'].fillna(0)
data_cancer['STDs'] = data_cancer['STDs'].fillna(1)
data_cancer['STDs (number)'] = data_cancer['STDs (number)'].fillna(data_cancer['STDs (number)'].median())
data_cancer['STDs:condylomatosis'] = data_cancer['STDs:condylomatosis'].fillna(data_cancer['STDs:condylomatosis'].median())
data_cancer['STDs:cervical condylomatosis'] = data_cancer['STDs:cervical condylomatosis'].fillna(data_cancer['STDs:cervical condylomatosis'].median())
data_cancer['STDs:vaginal condylomatosis'] = data_cancer['STDs:vaginal condylomatosis'].fillna(data_cancer['STDs:vaginal condylomatosis'].median())
data_cancer['STDs:vulvo-perineal condylomatosis'] = data_cancer['STDs:vulvo-perineal condylomatosis'].fillna(data_cancer['STDs:vulvo-perineal condylomatosis'].median())
data_cancer['STDs:syphilis'] = data_cancer['STDs:syphilis'].fillna(data_cancer['STDs:syphilis'].median())
data_cancer['STDs:pelvic inflammatory disease'] = data_cancer['STDs:pelvic inflammatory disease'].fillna(data_cancer['STDs:pelvic inflammatory disease'].median())
data_cancer['STDs:genital herpes'] = data_cancer['STDs:genital herpes'].fillna(data_cancer['STDs:genital herpes'].median())
data_cancer['STDs:molluscum contagiosum'] = data_cancer['STDs:molluscum contagiosum'].fillna(data_cancer['STDs:molluscum contagiosum'].median())
data_cancer['STDs:AIDS'] = data_cancer['STDs:AIDS'].fillna(data_cancer['STDs:AIDS'].median())
data_cancer['STDs:HIV'] = data_cancer['STDs:HIV'].fillna(data_cancer['STDs:HIV'].median())
data_cancer['STDs:Hepatitis B'] = data_cancer['STDs:Hepatitis B'].fillna(data_cancer['STDs:Hepatitis B'].median())
data_cancer['STDs:HPV'] = data_cancer['STDs:HPV'].fillna(data_cancer['STDs:HPV'].median())
data_cancer['STDs: Time since first diagnosis'] = data_cancer['STDs: Time since first diagnosis'].fillna(data_cancer['STDs: Time since first diagnosis'].median())
data_cancer['STDs: Time since last diagnosis'] = data_cancer['STDs: Time since last diagnosis'].fillna(data_cancer['STDs: Time since last diagnosis'].median())
```

```
cor_mat = data_cancer.corr()
plt.figure(figsize=(10,10))
sns.heatmap(cor_mat.astype(float),annot=True)
```

```
labels = data_cancer['Biopsy'].values.astype(float)
features = data_cancer.drop('Biopsy', axis=1).values.astype(float)
```

```
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```
X_train, X_test, y_train, y_test = train_test_split(features, labels,
                                                    test_size=0.25)
```

```
# Number of trees in the forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]

# number of features to consider when looking for the best split
max_features = ['auto', 'sqrt']

# Maximum depth of the tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)

# Minimum number of samples required to split an internal node
min_samples_split = [2, 5, 10]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]

# Method of selecting samples for training each tree
bootstrap = [True, False]

grid = {'n_estimators': n_estimators,
        'max_features': max_features,
        'max_depth': max_depth,
        'min_samples_split': min_samples_split,
        'min_samples_leaf': min_samples_leaf,
        'bootstrap': bootstrap}

print('Random grid :', grid)
```

```
rf = RandomForestClassifier()
rf_random = RandomizedSearchCV(estimator=rf,param_distributions=grid,
                               random_state=42)
rf_random.fit(X_train,y_train)
```

```
print("Best score with random forest classifier :", rf_random.best_score_)
print("Best params : ", rf_random.best_params_)
```

```
rf_random.score(X_test, y_test)
y_pred = rf_random.predict(X_test)

print(classification_report(y_test,y_pred))
```

```
pd.DataFrame(confusion_matrix(y_test,y_pred),
             columns=['Predicted No', 'Predicted Yes'],
             index=['Actual No', 'Actual Yes'])
```

```
scores = cross_val_score(rf_random, features, labels, scoring='f1_macro', cv=10)
print('Macro-F1 average: {0}'.format(scores.mean()))
```

```
kernel = ['rbf','poly','linear']
C = [0.1,1,10,100,1000]
gamma = [0.0001, 0.001, 0.01, 0.1, 1]

grid = {'kernel':kernel,
        'C':C,
        'gamma':gamma}

print("SVM param grid:", grid)
```

```
svc = SVC()
random_svc = RandomizedSearchCV(svc,param_distributions=grid,random_state=42)
random_svc.fit(X_train,y_train)
```

```
print("SVC best score:", random_svc.best_score_)
print("SVC best params:", random_svc.best_params_)
```

```
random_svc.score(X_test, y_test)
```

```
y_pred = random_svc.predict(X_test)
print(classification_report(y_test,y_pred))
```

```
pd.DataFrame(confusion_matrix(y_test,y_pred),
             columns=['Predicted No', 'Predicted Yes'],
             index=['Actual No', 'Actual Yes'])
```

```
scores = cross_val_score(random_svc,features,labels,scoring='f1_macro',cv=10)
print('Macro F1 average {0}'.format(scores.mean()))
```

```
from keras.models import Sequential
from keras.layers import Dense, Activation, BatchNormalization, Dropout
```

```
model = Sequential()

model.add(Dense(512,input_dim=35,activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))


model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',metrics=['accuracy'], 
              optimizer='adamax')
```

```
history = model.fit(X_train,y_train,validation_split=0.2,epochs=30, 
                    batch_size=100,verbose=2)
```

```
plt.title('Loss')
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])
plt.legend(['train', 'validation'],loc='best')
plt.xlabel('Epochs')

plt.show()

plt.title('Accuracy')
plt.plot(history.history['val_acc'])
plt.plot(history.history['acc'])
plt.legend(['train', 'validation'],loc='best')
plt.xlabel('Epochs')
plt.show()
```

```
model.evaluate(X_test,y_test)
```

```
y_pred = model.predict_classes(X_test)
print(classification_report(y_test, y_pred))
```

```
pd.DataFrame(confusion_matrix(y_test,y_pred),
            columns=['Predicted Yes','Predicted No'],
            index=['Actual Yes', 'Actual No'])
```

```
from xgboost import XGBClassifier

gamma = [0.5,1,1.5,2,5]
colsample_bytree = [0.6,0.8,1]
max_depth = [3,4,5,6,7]
min_child_weight = [1,5,10,20]
subsample = [0.6,0.8,1]


param_grid = {'gamma':gamma,
             'colsample_bytree':colsample_bytree,
             'max_depth':max_depth,
             'min_child_weight':min_child_weight,
             'subsample':subsample}

print("Parameters grid for XGBoost :", param_grid)


xgb = XGBClassifier()
random_xgb = RandomizedSearchCV(xgb,param_distributions=param_grid,
                                n_jobs=-1,random_state=42)
random_xgb.fit(X_train, y_train)

print("Best parameters :", random_xgb.best_params_)
print("Best score:", random_xgb.best_score_)
print("Accuracy on test set:",random_xgb.score(X_test, y_test))
```

```
y_pred = random_xgb.predict(X_test)
print(classification_report(y_test, y_pred))
```

```
pd.DataFrame(confusion_matrix(y_test, y_pred),
            columns=['Predicted Yes', 'Predicted No'],
            index=['Actual Yes','Actual No'])
```

```
scores = cross_val_score(random_xgb, features, labels, scoring='f1_macro',cv=10)
print("Macro-F1 average {0}".format(scores.mean()))
```
